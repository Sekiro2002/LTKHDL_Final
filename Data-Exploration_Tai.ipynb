```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_csv('archive/data.csv', delimiter = ';')
df_1
```

```python
df_2 = pd.read_csv('archive/update.csv', delimiter = ';' )
df_2
```

```python
df = pd.concat([df_1, df_2], ignore_index = True)
df
```

## 1. Điểm trung bình từng môn 

### ý nghĩa: cho biết điểm trung bình của từng môn thi trong đợt thi trung học phổ thông 2021


### Xóa các giá trị Nan

```python
math = df['math'].dropna()
physics = df['physics'].dropna()
chemistry = df['chemistry'].dropna()
biology = df['biology'].dropna()
history = df['history'].dropna()
geography = df['geography'].dropna()
literature = df['literature'].dropna()
civic_education = df['civic_education'].dropna()
english = df['english'].dropna()
```

### Tính điểm trung bình 

```python
math_avg = round(np.average(math), 2)
physics_avg = round(np.average(physics), 2)
chemistry_avg = round(np.average(chemistry), 2)
biology_avg = round(np.average(biology), 2)
history_avg = round(np.average(history), 2) 
geography_avg = round(np.average(geography), 2)
literature_avg = round(np.average(literature), 2)
civic_education_avg = round(np.average(civic_education), 2)
english_avg = round(np.average(english), 2)
```

### Trực quan hóa

```python
columns_name = ["Math", "Literature", "English", "Physics", "Chemistry", "Biology", "History",
                    "Geography", "Civic Education"]
values = [math_avg, literature_avg, english_avg, physics_avg, chemistry_avg, biology_avg, history_avg, 
                      geography_avg, civic_education_avg]
fig, ax = plt.subplots(figsize = (18, 12))
sns.set(font_scale = 1.35)
score_average = sns.barplot(x = columns_name, y = values)

for counts_10 in score_average.patches:
  height = counts_10.get_height()
  label_x = counts_10.get_x() + counts_10.get_width() / 2
  label_y = counts_10.get_y() / height + height
  plt.text(label_x, label_y, s = f"{height:}", ha='center', va='bottom', color = "black", size = 15)

plt.xlabel("Subjects")
plt.ylabel("Occurences")
plt.title("GPA for each subjects");
```

### Nhận xét:
- Các môn thi trừ môn GDCD đều có điểm thi trung bình dưới mức 7. 
- Các môn Lịch sử có điểm trung bình dưới trung bình -> Theo đánh giá thì đề thi môn Lịch sử năm nay khá khó.

### b. Ngữ văn

#### Đếm số lượng của từng số điểm thi và xử lí các số điểm thi bị lẻ bằng cách cộng số lượng của điểm đó vào điểm số gần nhất

```python
literature_df = df['literature'].value_counts().sort_index()
x = np.arange(0,10.1,0.25)
for v in literature_df.index:
    if (v not in x):
        literature_df[round(v/0.25)*0.25] += literature_df[v] 
        literature_df = literature_df.drop(index = v)
literature_df.sum()
```

```python
fig, ax = plt.subplots(figsize = (25, 15))
ax = sns.barplot(x = literature_df.index, y = literature_df.values)

for counts_10 in ax.patches:
  height = counts_10.get_height()
  label_x = counts_10.get_x() + counts_10.get_width() / 2
  label_y = counts_10.get_y() / height + height
  plt.text(label_x, label_y, s = f"{int(height):}", ha='center', va='bottom', color = "black", size = 14)

plt.xlabel('Điểm', fontsize = 25)
plt.ylabel('Số thí sinh', fontsize = 25)
plt.title('Biểu đồ phân bố điểm thi THPT Quốc Gia môn Ngữ Văn năm 2021', fontsize = 30);
```

```python
fig, ax = plt.subplots(figsize = (25, 15))
ax = sns.barplot(x = japanese_df.index, y = japanese_df.values)
#ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
for counts_10 in ax.patches:
  height = counts_10.get_height()
  label_x = counts_10.get_x() + counts_10.get_width() / 2
  label_y = counts_10.get_y() / height + height
  plt.text(label_x, label_y, s = f"{int(height):}", ha='center', va='bottom', color = "black", size = 15)

plt.xlabel('Điểm', fontsize = 25)
plt.ylabel('Số thí sinh', fontsize = 25)
plt.title('Biểu đồ phân bố điểm thi THPT Quốc Gia môn Tiếng Nhật năm 2021', fontsize = 30);
```
